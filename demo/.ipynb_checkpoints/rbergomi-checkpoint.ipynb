Rough volatility for FX (GBPUSD)
==

The starting point for our rough FX model is rBergomi written like this

$$S_t = \mathcal{E} \left\{ \int_0^t \sqrt{ v_s } \mathrm{d}Z_s \right\}, $$

$$v_t = \mathcal{E} \left\{ \eta \sqrt{2\alpha + 1} \int_0^t (t - s)^\alpha \left( \rho \mathrm{d}Z_s + \sqrt{1 - \rho^2} \mathrm{d}W_s \right) \right\}, \quad \left[ Z, W \right]_t = 0.$$

Forward curves are neglected, calibrations to which will remain unchanged. Separating integrators, we have 

$$v_t = \mathcal{E} \left\{ \rho \eta \sqrt{2\alpha + 1} \int_0^t (t - s)^\alpha \mathrm{d}Z_s + \sqrt{1 - \rho^2} \eta \sqrt{2\alpha + 1} \int_0^t (t - s)^\alpha \mathrm{d}W_s \right\}.$$

Now let truncated BSS processes be defined as

$$Y_t^\alpha := \sqrt{2\alpha + 1} \int_0^t (t - s)^\alpha \mathrm{d}Y_s.$$

The process $Z_t^\alpha$ is responsible for implied volatility skew; the process $W_t^\alpha$ is responsible for implied volatility smile.

This rBergomi model applied to SPX data appears to work well, *because skew and smile appear to explode and relax over similar time-scales*. Equivalently, for SPX, it's OK that $Z_t^\alpha$ and $W_t^\alpha$ share the same roughness index.

This constraint appears inhibitive for FX. Although smile is explosive in short-time (very well documented), skew is not; short-time implied volatilities are much more symmetric than for SPX.

To control the explosion and relaxation of skew and smile for FX independently, consider decoupling their roughness indices

$$X_t = \mathcal{E} \left\{ \int_0^t \sqrt{ v_s } \mathrm{d}Z_s \right\}, \quad v_t = \mathcal{E} \left\{ \rho Z^\beta_t + \eta W^\alpha_s \right\} $$

where $\rho$ and $\eta$ are redefined (one can only really relate these parameters to the rBergomi equivalent for a single time thanks to decoupled roughness indices, so better to redefine them as opposed to continuing with $\rho\eta$ and $\sqrt{1 - \rho^2}\eta$). $\rho$ and $\eta$ remind us which BSS process predominantly controls skew and smile.

Initial tests suggest that FX skew is not at all explosive in short-time, and in fact the newly introduced roughness index $\beta$ can be relaxed all the way to zero. With FX not being exposed to leverage effects, perhaps this should not be surprising.

The model demonstrated below, and compared with rBergomi, therefore takes the full form

$$X_t = \mathcal{E} \left\{ \int_0^t \sqrt{ v_s } \mathrm{d}Z_s \right\}, $$

$$v_t = \mathcal{E} \left\{ \rho Z_t + \eta \sqrt{2\alpha + 1} \int_0^t (t - s)^\alpha \mathrm{d}W_s \right\}, \quad \left[ Z, W \right]_t = 0.$$

Import required libraries and classes

```python
import numpy as np
from matplotlib import pyplot as plt
from rbergomi import rBergomi
from surface import Surface
from calibrate import Calibrate
% matplotlib inline
```

Import data for calibration. Appropriate Excel file should be in rbergomi/data

```python
surface = Surface('GBPUSD', '20161230', close = False)
```

Display Pandas surface

```python
surface.surface
```

Create instance of the rBergomi class with granularity $n \times 156$, paths $N$, max maturity $T$ and roughness index $a$

```python
rbergomi = rBergomi(n = 8, N = 40000, T = 1.0, a = -0.43, AV = True)
```

Bootstrap piecewise flat instantaneous forward variance, $\xi(t)$, from implied integrated variances in data. Discretising as required for rBergomi instance

```python
xi = surface.xi(n = rbergomi.n)
```

Demonstrate $\xi(t)$ against the rBergomi time grid

```python
t = rbergomi.t 
plot, axes = plt.subplots()
axes.plot(t[0,:], xi[0,:])
plt.grid(True)
```

Fix the generator's seed for replicable simulation results

```python
np.random.seed(0)
```

Generate required orthogonal random numbers. Recall $W_1$ is mulivariate (required for the hybrid scheme), depending on $a$

```python
dW1 = rbergomi.dW1()
dW2 = rbergomi.dW2()
```

Construct the truncated BSS process, $Y(t)$

```python
Y = rbergomi.Y1(dW1)
```

Now correlate the orthogonal increments. 1 will be used for rBergomi, 2 for rFX (hence original $\rho = 0$).

```python
dZ1 = rbergomi.dZ(dW1, dW2, rho = -0.4) 
dZ2 = rbergomi.dZ(dW1, dW2, rho = +0.0) 
```

Construct the variance process, $v(t)$ using piecewise flat $\xi(t)$

```python
v1 = xi * rbergomi.v1(Y, eta = 1.6) 
v2 = xi * rbergomi.v1(Y, eta = 1.5) 
```

Finally the rBergomi and rFX price process, $S(t)$

```python
S1 = rbergomi.S(v1, dZ1)
```

```python
S2 = rbergomi.X(v2, dZ2, eta1 = -0.64)
```

Construct call price surface

```python
C1 = rbergomi.C(S1, surface)
C2 = rbergomi.C(S2, surface)
```

converting these to implied volatilities

```python
IV1 = rbergomi.IV(C1, surface)
IV2 = rbergomi.IV(C2, surface)
```

Finally, plot results against the market data

```python
k = np.array(surface._log_strike_surface())
MV = np.array(surface.surface)

for M in range(len(k[:,0])):
    plot, axes = plt.subplots()

    # axes.plot(k[M,:], IV[M,:,0], 'k--', linewidth = 0.5) # Lower bound
    axes.plot(k[M,:], IV1[M,:,1], 'r')
    axes.plot(k[M,:], IV2[M,:,1], 'b')
    # axes.plot(k[M,:], IV[M,:,2], 'k--', linewidth = 0.5) # Upper bound

    axes.plot(k[M,:], MV[M,:], 'go', fillstyle = 'none', ms = 4, mew = 1)
    
    axes.set_xlabel(r'$k$', fontsize = 14)
    axes.set_ylabel(r'$IV(k,T)$', fontsize = 14)
    axes.set_title(r'$T=$' + surface._tenors[M], fontsize = 14)
    axes.legend([r'$\mathsf{rBergomi}$', r'$\mathsf{rFX}$', r'$\mathsf{Market}$'])
    plt.grid(True)
```

It is clear to see (for this instance of FX data) that the term structure of skew implied by rBergomi is not good; FX skew does not explode as $T \to 0$, in fact, skew appears to tend to 0 itself.

The remainder checks quality of convergence. Demonstrate antithetic paths

```python
t = rbergomi.t
paths = rbergomi.N * np.array([0,1,2,3])
plot, axes = plt.subplots()
for p in paths:
    axes.plot(t[0,:], S2[p,:])
```

Test expectation of price process paths using $S_0(t) = \mathbb{E}^\mathbb{Q}_0[S(t)]$

```python
plot, axes = plt.subplots()
mean = np.mean(S2, axis = 0)
std = np.std(S2, axis = 0)/np.sqrt(len(S2[:,0]))
axes.plot(t[0,:], mean, 'r')
axes.plot(t[0,:], mean + 2 * std, 'k--', linewidth = 0.5)
axes.plot(t[0,:], mean - 2 * std, 'k--', linewidth = 0.5)
axes.set_xlabel(r'$t$', fontsize = 14)
axes.legend([r'$\mathbb{E}^\mathbb{Q}_0[S(t)]$', r'$95\%$'])
axes.set_ylim([0.99, 1.01])
plt.grid(True)
```

Test expectation of variance process using $\xi_0(t) = \mathbb{E}^\mathbb{Q}_0[v(t)]$. Current variance process not returned for rFX, hence showing rBergomi

```python
plot, axes = plt.subplots()
axes.plot(t[0,:], xi[0,:], 'b')
axes.plot(t[0,:], np.mean(v1, axis = 0), 'r')
axes.legend([r'$\xi_0(t)$', r'$\mathbb{E}^\mathbb{Q}_0[v(t)]$'])
axes.set_xlabel(r'$t$', fontsize = 14)
plt.grid(True)
```

Check that the price process has inherited intended variance using $\int_0^t \xi_0(s)\mathrm{d}s = -2\mathbb{E}^\mathbb{Q}_0[\log{S_t}]$

```python
std1 = np.cumsum(xi[0,:-1]) * rbergomi.dt 
std2 = -2 * np.mean(np.log(S2[:,1:]), axis = 0)
plot, axes = plt.subplots()
axes.plot(t[0,1:], std1, 'b')
axes.plot(t[0,1:], std2, 'r')
axes.set_xlabel(r'$t$', fontsize = 14)
plt.grid(True)
```
