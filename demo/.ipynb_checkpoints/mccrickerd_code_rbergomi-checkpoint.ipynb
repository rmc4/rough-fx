Pricing under rough volatility demonstration
==

I demonstrate the following rBergomi model using $\eta = 1.9$, $\rho = -0.9$ and $H = \alpha + \frac{1}{2} = 0.07 \ $ from Bayer *et al.* (2016)

$$S_t = \mathcal{E} \left\{ \int_0^t \sqrt{ v_s } \mathrm{d}Z_s \right\}, $$

$$v_t = \mathcal{E} \left\{ \eta \sqrt{2\alpha + 1} \int_0^t (t - s)^\alpha \mathrm{d}W_s \right\}, \quad \left[ Z, W \right]_t = \rho t.$$

Change directory to folder with rBergomi files. All files used here can be found at https://github.com/ryanmccrickerd/rbergomi

```python
import os
os.chdir('/Users/ryanmccrickerd/desktop/rbergomi/rbergomi')
```

Import required libraries and classes

```python
import numpy as np
from matplotlib import pyplot as plt
from rbergomi import rBergomi
from surface import Surface
from calibrate import Calibrate
% matplotlib inline
```

Import implied mid volatility data. This is from the same date as used in their paper, although from Bloomberg so inevitably different

```python
surface = Surface('SPX', '20100204', close = True)
```

Display surface

```python
surface.surface
```

Create instance of the rBergomi class with granularity $n \times 156$, paths $N$, max maturity $T$ and roughness index $a$

```python
rbergomi = rBergomi(n = 8, N = 100000, T = 1.0, a = -0.43, AV = True)
```

Bootstrap piecewise flat instantaneous forward variance, $\xi(t)$, from implied integrated variances in implied volatility data. Discretising as required for rBergomi instance

```python
xi = surface.xi(n = rbergomi.n)
```

Demonstrate $\xi(t)$ against the rBergomi time grid

```python
t = rbergomi.t 
plot, axes = plt.subplots()
axes.plot(t[0,:], xi[0,:])
plt.grid(True)
```

Fix the generator's seed for replicable simulation results

```python
np.random.seed(0)
```

Generate required orthogonal random numbers. $W_1$ is mulivariate (using the hybrid scheme), and depends on $a$

```python
dW1 = rbergomi.dW1()
dW2 = rbergomi.dW2()
```

Construct the required Volterra process, $Y(t)$

```python
Y = rbergomi.Y1(dW1)
```

Now correlate the orthogonal increments appropriately

```python
dZ = rbergomi.dZ(dW1, dW2, rho = -0.9) 
```

Construct the variance process, $v(t)$ using piecewise flat $\xi(t)$

```python
v = xi * rbergomi.v1(Y, eta = 1.9) 
```

Finally build the rBergomi price process, $S(t)$

```python
S = rbergomi.S(v, dZ)
```

Construct call price surface for points in data

```python
C = rbergomi.C(S, surface)
```

converting these to implied volatilities

```python
IV = rbergomi.IV(C, surface)
```

Finally, plot results against the data

```python
k = np.array(surface._log_strike_surface()) # Log-strike surface
MV = np.array(surface.surface) # Market vols

for M in range(len(k[:,0])):
    plot, axes = plt.subplots()
    
    axes.plot(k[M,:], IV[M,:,1], 'r')
    axes.plot(k[M,:], MV[M,:], 'bo', fillstyle = 'none', ms = 4, mew = 1)    
    
    axes.plot(k[M,:], IV[M,:,0], 'k--', linewidth = 0.5) # 1sd lower bound
    axes.plot(k[M,:], IV[M,:,2], 'k--', linewidth = 0.5) # 1sd upper bound

    axes.set_xlabel(r'$k$', fontsize = 14)
    axes.set_ylabel(r'$IV(k,T)$', fontsize = 14)
    axes.set_title(r'$T=$' + surface._tenors[M], fontsize = 14)
    axes.legend([r'$\mathsf{rBergomi}$', r'$\mathsf{Market}$'])
    plt.grid(True)
```

It is clear to see that the parameters obtained in their paper do not match my data as well, but generally the fit is still reasonable. Putting data aside, my rBergomi implied volatilities match theirs.

The remainder checks quality of convergence. Demonstrate price and volatility antithetic paths

```python
t = rbergomi.t
paths = rbergomi.N * np.array([0,1,2,3])
plot, axes = plt.subplots()
for p in paths:
    axes.plot(t[0,:], S[p,:])
```

Test expectation of price process paths using $1 = S_0(t) = \mathbb{E}^\mathbb{Q}_0[S(t)]$

```python
plot, axes = plt.subplots()
mean = np.mean(S, axis = 0)
std = np.std(S, axis = 0)/np.sqrt(len(S[:,0]))
axes.plot(t[0,:], mean, 'r')
axes.plot(t[0,:], mean + 2 * std, 'k--', linewidth = 0.5)
axes.plot(t[0,:], mean - 2 * std, 'k--', linewidth = 0.5)
axes.set_xlabel(r'$t$', fontsize = 14)
axes.legend([r'$\mathbb{E}^\mathbb{Q}_0[S(t)]$', r'$95\%$'])
axes.set_ylim([0.99, 1.01])
plt.grid(True)
```

Test expectation of variance process using $\xi_0(t) = \mathbb{E}^\mathbb{Q}_0[v(t)]$

```python
plot, axes = plt.subplots()
axes.plot(t[0,:], xi[0,:], 'b')
axes.plot(t[0,:], np.mean(v, axis = 0), 'r', linewidth = 0.5)
axes.legend([r'$\xi_0(t)$', r'$\mathbb{E}^\mathbb{Q}_0[v(t)]$'])
axes.set_xlabel(r'$t$', fontsize = 14)
plt.grid(True)
```

Check that the price process has inherited intended variance using $\int_0^t \xi_0(s)\mathrm{d}s = -2\mathbb{E}^\mathbb{Q}_0[\log{S_t}]$

```python
std1 = np.cumsum(xi[0,:-1]) * rbergomi.dt 
std2 = -2 * np.mean(np.log(S[:,1:]), axis = 0)
plot, axes = plt.subplots()
axes.plot(t[0,1:], std1, 'b')
axes.plot(t[0,1:], std2, 'r')
axes.set_xlabel(r'$t$', fontsize = 14)
plt.grid(True)
```
