Change directory to folder with rBergomi scripts

Import required libraries and classes, including rBergomi, Surface and Calibrate

```python
import numpy as np
from matplotlib import pyplot as plt
from rbergomi import rBergomi
from surface import Surface
from surface_utils import xl_import
from calibrate import Calibrate
% matplotlib inline
```

Import data for calibration. Appropriate Excel file should be in rbergomi/data

```python
date = '20170206'
index = 'SPX'
data = xl_import(index, date, close = False)
surface = Surface(data)
```

Display Pandas surface

```python
surface.surface
```

Create instance of the rBergomi class with granularity $n \times 156$, paths $N$, max maturity $T$ and roughness index $a$

```python
rbergomi = rBergomi(n = 2, N = 2500, T = 1.0, a = -0.41)
```

Bootstrap piecewise flat instantaneous forward variance, $\xi(t)$, from implied integrated variances in data. Discretising as required for rBergomi instance

```python
xi = surface.flat_xi(n = rbergomi._n/156.) # 'orrible
```

Demonstrate $\xi(t)$ against the rBergomi time grid

```python
t = rbergomi._t # These are not private. So remove underscores
plot, axes = plt.subplots()
axes.plot(t, xi)
```

Fix the generator's seed for replicable results

```python
np.random.seed(0)
```

Generate required orthogonal random numbers. Recall $W_1$ is mulivariate, depending on $a$

```python
dW1 = rbergomi.dW1()
dW2 = rbergomi.dW2()
```

Construct the truncated BSS process, $Y(t)$

```python
Y = rbergomi.Y(dW1)
```

*Everything up to this point does not require re-running in a calibration of $\rho$ and $\eta$*. Now correlate orthogonal increments

```python
dZ = rbergomi.dZ(dW1, dW2, rho = -0.906393551143)
```

Construct the price process, $S(t)$, via the variance process, $v(t)$. Now include piecewise constant $\xi(t)$

```python
v = xi * rbergomi.v(Y, eta = 1.93579401161)
```

*Constructing this price process next is the slowest part*

```python
S = rbergomi.S(v, dZ)
```

Demonstrate $2^{\mathrm{factors}}$ antithetic paths

```python
t = rbergomi._t
p = rbergomi._N * np.array([0,1,2,3])
plot, axes = plt.subplots()
for i in p:
    axes.plot(t, S[:,i])
```

Assign requirements for rBergomi implied surface

```python
M = surface._maturities
k = np.array(surface._log_strike_surface())
```

Build rBergomi surface for all maturities and strikes in data

```python
rbergomi_surface = rbergomi.surface(S, M, k)
```

Demonstrate these

```python
implied = rbergomi_surface
plot, axes = plt.subplots()
for i in range(5): # Generalise
    axes.plot(k[i,:], implied[i,:])
```

Also show the actual surface

```python
actual = np.array(surface.surface)
plot, axes = plt.subplots()
for i in range(5): # Generalise
    axes.plot(k[i,:], actual[i,:])
```

Show RMSE of actual vs. implied surfaces for these parameters (Bayer 2015)

```python
rbergomi.rmse(actual, implied)
```

Set up calibration object for the chosen seed

```python
calibrate = Calibrate(rbergomi, surface, seed = 0)
```

Run calibration with L-BFGS-B solver

```python
calibrate.run(method = 'L-BFGS-B', maxiter = 10, rho0 = -0.9, eta0 = 1.9)
```
