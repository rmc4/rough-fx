Import required libraries and classes

```python
import numpy as np
from matplotlib import pyplot as plt
from rbergomi import rBergomi
from surface import Surface
from calibrate import Calibrate
% matplotlib inline
```

Create instance of the rBergomi class with granularity $n \times 156$, paths $N$, max maturity $T$ and roughness index $a$

```python
n = 2
N = 4 * 1
a = 0.43
```

```python
rbergomiA = rBergomi(n = n, N = N, T = 1.0, a = + a, AS = True)
rbergomiB = rBergomi(n = n, N = N, T = 1.0, a = 0.0000001, AS = True)
rbergomiC = rBergomi(n = n, N = N, T = 1.0, a = - a, AS = True)
```

Fix the generator's seed for replicable simulation results

Generate required orthogonal random numbers. Recall $W_1$ is mulivariate (required for the hybrid scheme), depending on $a$

USE 400 HERE FOR NICE RESULTS

```python
seed = 400
```

```python
np.random.seed(seed)
dW1A = rbergomiA.dW1()
dW2A = rbergomiA.dW2()
np.random.seed(seed)
dW1B = rbergomiB.dW1()
dW2B = rbergomiB.dW2()
np.random.seed(seed)
dW1C = rbergomiC.dW1()
dW2C = rbergomiC.dW2()
```

Construct the truncated BSS process, $Y(t)$

```python
YA = rbergomiA.Y(dW1A)
YB = rbergomiB.Y(dW1B)
YC = rbergomiC.Y(dW1C)
```

*Now demonstrate this!*

```python
t = rbergomiA.t
dpi = 10000
figsize = (4,2.5)
plot_paths = 2
ylim = 3.
```

```python
plot, axes = plt.subplots(figsize=figsize)
axes.plot(t[0,:], +YB[0,:], 'b')
axes.plot(t[0,:], -YB[0,:], 'r', linewidth = 0.25)
# axes.plot(t[0,:], +YB[1,:], 'r')
# axes.plot(t[0,:], -YB[1,:], 'r', linewidth = 0.5)
axes.set_xlabel(r'$t$', fontsize = 16)
axes.set_ylabel(r'$W_t^\alpha$', fontsize = 16)
axes.set_title(r'$\alpha = 0$', fontsize = 16)
# axes.legend([r'$\alpha = -0.00$'])
plt.ylim([-ylim,ylim])
plt.grid(True)

fname = '/Users/ryanmccrickerd/Desktop/phd/2017-18/turbo/volterra0.pdf'
plt.savefig(fname, dpi=300, bbox_inches='tight')
```

```python
plot, axes = plt.subplots(figsize=figsize)
axes.plot(t[0,:], +YC[0,:], 'b')
axes.plot(t[0,:], -YC[0,:], 'r', linewidth = 0.25)
# axes.plot(t[0,:], +YC[1,:], 'r')
# axes.plot(t[0,:], -YC[1,:], 'r', linewidth = 0.5)
axes.set_xlabel(r'$t$', fontsize = 16)
axes.set_ylabel(r'$W_t^\alpha$', fontsize = 16)
axes.set_title(r'$\alpha = -0.43$', fontsize = 16)
# axes.legend([r'$\alpha = -0.43$'])
plt.ylim([-ylim,ylim])
plt.grid(True)

fname = '/Users/ryanmccrickerd/Desktop/phd/2017-18/turbo/volterra1.pdf'
plt.savefig(fname, dpi=300, bbox_inches='tight')
```

```python
dBB = rbergomiB.dB(dW1B, dW2B, rho = -0.0001)
dBC = rbergomiC.dB(dW1C, dW2C, rho = -0.0001) 
```

Construct the variance process, $v(t)$ using piecewise flat $\xi(t)$

```python
VB = rbergomiB.V(YB, xi = 0.235**2, eta = 1.9) 
VC = rbergomiC.V(YC, xi = 0.235**2, eta = 1.9) 
```

Finally the rBergomi and rFX price process, $S(t)$

```python
SB = rbergomiB.S(VB, dBB)
SC = rbergomiC.S(VC, dBC)
```

```python
plot, axes = plt.subplots(figsize=figsize)
axes.plot(t[0,:], SC[0,:], 'b')
axes.plot(t[0,:], SC[1,:], 'b', linewidth = 0.25)
axes.plot(t[0,:], SC[2,:], 'r')
axes.plot(t[0,:], SC[3,:], 'r', linewidth = 0.25)
# axes.plot(t[0,:], -YC[0,:], 'b', linewidth = 0.25)
# axes.plot(t[0,:], +YC[1,:], 'r')
# axes.plot(t[0,:], -YC[1,:], 'r', linewidth = 0.5)
axes.set_xlabel(r'$t$', fontsize = 16)
axes.set_ylabel(r'$S_t$', fontsize = 16)
axes.set_title(r'$\rho = 0,\ \alpha = -0.43$', fontsize = 16)
# axes.legend([r'$\alpha = -0.43$'])
plt.ylim([0.7,1.3])
plt.grid(True)

# fname = '/Users/ryanmccrickerd/Desktop/phd/2017-18/turbo/price0.pdf'
fname = '/Users/ryanmccrickerd/Desktop/phd/2017-18/turbo/price1.pdf'
plt.savefig(fname, dpi=300, bbox_inches='tight')
```
