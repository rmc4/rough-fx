We have: $\Gamma(z, a, b) = \int_a^b t^{z - 1}e^{-t} \mathrm{d}t$

We want: $\int_0^t x^{2a} e^{-2bx} \mathrm{d}x$

```python
a = -0.3
import numpy as np
t = np.linspace(0, 1, 100 + 1)
```

```python
t
```

```python
from matplotlib import pyplot as plt
% matplotlib inline
plot, axes = plt.subplots()
axes.plot(t, t**(2*a + 1), 'b', linewidth = 1)
```

```python
t = np.linspace(0, 1, 100 + 1)
integral = np.zeros_like(t)
```

```python
alpha = -0.3
beta = 1.0

for i in np.arange(1, len(integral), 1):
    integral[i] = mpmath.gammainc(2*alpha + 1, a = 0, b = 2*beta*t[i], regularized = False)

#gamma = np.vectorize(mpmath.gammainc)    
#integral = gamma(2*alpha + 1, a = 0, b = 2*beta*t, regularized = False)    
    
#integral = integral / (2*beta)**(2*alpha + 1)
# Rescale to pass through (1,1) - should remove regularised token above if going to use this
scale = (2*alpha + 1) / (2*beta)**(2*alpha + 1)
integral = scale * integral
    
plot, axes = plt.subplots()
axes.plot(t, integral, 'b', linewidth = 1)
axes.plot(t, t**(2*a + 1), '--r', linewidth = 1)
```

```python
t = np.linspace(0, 1, 100 + 1)
ratio = np.zeros_like(t)
t2 = 1
```

```python
alpha = -0.3
beta = 1 # Doesn't matter since reformulated below

for i in np.arange(1, len(integral), 1):
    ratio[i] = t[i]**(2*alpha + 1) / mpmath.gammainc(2*alpha + 1, b = t[i]*t2, regularized = False)

ratio[0] = None
    
plot, axes = plt.subplots()
axes.plot(t, np.sqrt(ratio), 'b', linewidth = 1)
axes.plot(t, np.sqrt(2*a + 1)*np.ones_like(t), 'k', linewidth = 1)
```
